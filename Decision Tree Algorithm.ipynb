# Decision Tree Algorithm
# Prediction of the possibility of diabetes
# based on some features

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier # or DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('E:\Momeni-Machne Learning class momeni 21 01 1402\Machine Learning\dataset/diabetes.csv')
data
```

```python
x= pd.DataFrame(data, columns = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'])
y=data.Outcome.values.reshape(-1,1) # Target Variable
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
clf=DecisionTreeClassifier(max_depth=3)
clf=clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
print('Accuracy:',metrics.accuracy_score(y_test,y_pred))
```

```python
# criterion: entropy
clf=DecisionTreeClassifier(criterion='entropy',max_depth=3)
clf=clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
print('Accuracy:',metrics.accuracy_score(y_test,y_pred))
```
